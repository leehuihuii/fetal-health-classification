# **Model Development**

Define function to print performance of model

---



````python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
````

````python
def performance(score,y_actual,y_pred):
  # print accuracy score
  print('Accuracy score :{}\n'.format(score))

  # print classification report
  print('Classification report:')
  target_names = ['Normal (0)','Suspect (1)','Pathological (2)']
  print(classification_report(y_actual,y_pred,target_names = target_names,digits=4))

  # print confusion matrix graph
  print('Confusion Matrix: ')
  cm = confusion_matrix(y_actual,y_pred)
  plt.figure(figsize=(8,6))
  sns.heatmap(cm,annot=True)
  plt.xlabel('Predicted')
  plt.ylabel('Actual')
````

`Learning curve`

Learning curves of model performance on the train and test datasets can be used to diagnose an underfit, overfit, or well-fit model.

````python
from sklearn.model_selection import learning_curve


def plot_learning_curve(model, title):
  train_sizes, train_scores, test_scores = learning_curve(model, X_train_sm, y_train_sm, cv=5, scoring='accuracy', train_sizes=np.linspace(0.1, 1.0, 10))

  train_scores_mean = np.mean(train_scores, axis=1)
  train_scores_std = np.std(train_scores, axis=1)
  test_scores_mean = np.mean(test_scores, axis=1)
  test_scores_std = np.std(test_scores, axis=1)

  fig = go.Figure()
  # fig.update_yaxes(range=[0.5, 1])
  fig.add_trace(go.Scatter(x=train_sizes,
                            y=train_scores_mean,
                            mode='lines',
                            name='Training Score'))

  fig.add_trace(go.Scatter(x=train_sizes,
                            y=test_scores_mean,
                            mode='lines',
                            name='Testing Score'))

  fig.add_trace(go.Scatter(x=np.concatenate([train_sizes, train_sizes[::-1]]),
                            y=np.concatenate([train_scores_mean - train_scores_std,
                                                (train_scores_mean + train_scores_std)[::-1]]),
                            fill='tozeroy',
                            mode='none',
                            name='Training Score Std Dev',
                            line=dict(color='rgba(255, 0, 0, 0.1)')))

  fig.add_trace(go.Scatter(x=np.concatenate([train_sizes, train_sizes[::-1]]),
                            y=np.concatenate([test_scores_mean - test_scores_std,
                                                (test_scores_mean + test_scores_std)[::-1]]),
                            fill='tozeroy',
                            mode='none',
                            name='Testing Score Std Dev',
                            line=dict(color='rgba(0, 255, 0, 0.1)')))

  fig.update_layout(title=title,
                      xaxis_title='Training Set Size',
                      yaxis_title='Score',
                      xaxis=dict(range=[850, 1700]),
                      yaxis=dict(range=[0.75, 1.05])
                      )

  fig.show()
````

````python
title = 'Learning curve (Decision Tree)'
plot_learning_curve(model_tree,title)
````

````python
title = 'Learning curve (Random Forest)'
plot_learning_curve(model_random,title)
````

````python
title = 'Learning curve (XGBoost Classifier)'
plot_learning_curve(model_xgbc,title)
````

````python
title = 'Learning curve (Logistic Regression)'
plot_learning_curve(model_lr,title)
````



> **Observation:**


```
In terms of the gap between training and testing performance line, the logistic regression performs the best as the gap is the smallest.
This indicates that the logistic regression model is not overfitting and is able to generalize well to unseen data.

For decision tree, the model is overfitting and not able to generalize well to unseen data as testing accuracy score start to decline when amount of data is higher than 1500.
```

`ROC Curve`

ROC curve stands for receiver operating characteristics curve, it is a graph showing the performance of a classification model at all classification thresholds. This curve plots two parameters which are TPR, the true positive rate and FPR, the false positive rate. TPR also known as sensitivity tells us that what proportion of positive class got correctly classified. FPR tells us what proportion of the negative class got incorrectly classified by the classifier.

````python
from sklearn.metrics import roc_curve, auc

def plot_roc_curve(model,title):
  # Get the predicted probabilities for each class
  y_scores = model.predict_proba(X_test)

  # Compute ROC curve and ROC area for each class
  n_classes = 3
  fpr = dict()
  tpr = dict()
  roc_auc = dict()
  for i in range(n_classes):
      fpr[i], tpr[i], _ = roc_curve(y_test, y_scores[:, i], pos_label=i)
      roc_auc[i] = auc(fpr[i], tpr[i])

  # Create the ROC curves using plotly.graph_objects
  fig = go.Figure()
  colors = ['blue', 'red', 'green']
  for i, color in zip(range(n_classes), colors):
      fig.add_trace(go.Scatter(x=fpr[i], y=tpr[i], mode='lines', name='Class {0} (AUC = {1:.2f})'.format(i, roc_auc[i]), line=dict(color=color, width=2)))

  fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], mode='lines', name='Random', line=dict(color='black', width=2, dash='dash')))
  fig.update_layout(
      title=title,
      xaxis=dict(title='False Positive Rate'),
      yaxis=dict(title='True Positive Rate'),
      showlegend=True,
      legend=dict(x=1, y=0)
  )
  fig.show()
````

````python
title = 'Receiver Operating Characteristic (ROC) Curve - Decision Tree'
plot_roc_curve(model_tree,title)
````

````python
title = 'Receiver Operating Characteristic (ROC) Curve - Random Forest Classifier'
plot_roc_curve(model_random,title)
````

````python
title = 'Receiver Operating Characteristic (ROC) Curve - XGBoost Classifier'
plot_roc_curve(model_xgbc,title)
````

````python
title = 'Receiver Operating Characteristic (ROC) Curve - Logistic Regression'
plot_roc_curve(model_lr,title)
````



> **Observations:**


```
To evaluate the performance of models, we look at the AUC, which is the area under the curve.
The larger the AUC, the better the model performing.
We have three classes in our classification problem. Thus, we take the average value of the AUC of class0, class1 and class2.
The XGBoost Classifier obtain the highest average AUC among the four models while decision tree has the lowest average AUC.
```
